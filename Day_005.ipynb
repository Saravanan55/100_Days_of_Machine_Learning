# **Threading Module**

```
import threading
from queue import Queue
import time

print_lock = threading.Lock()
q=Queue()
def Job(worker):
    time.sleep(.5) 
    with print_lock:
        print(threading.current_thread().name,worker)

def threader():
    while True:
        worker = q.get()
        Job(worker)
        q.task_done()

for x in range(5):
     t = threading.Thread(target=threader)
     t.daemon = True
     t.start()

start = time.time()
for worker in range(10):
    q.put(worker)
q.join()
print('Entire job took:',time.time() - start)
```

# **Matplotlib Styles**

```
from matplotlib import pyplot as p
from matplotlib import style
style.use('ggplot')
x=[5,6,7,8]
y=[7,3,8,3]
x2=[5,6,7,8]
y2=[6,7,2,6]
p.plot(x,y,"g",linewidth=5)
p.plot(x2,y2,"r",linewidth=10)

p.title("Epic Info")
p.xlabel("X axis")
p.ylabel("Y axis")
p.show()
```

# **Matplotlib Legends**

```
from matplotlib import pyplot as p
from matplotlib import style
#style.use('ggplot')
x=[5,6,7,8]
y=[7,3,8,3]
x2=[5,6,7,8]
y2=[6,7,2,6]
p.plot(x,y,"g",linewidth=5, label="Line one")
p.plot(x2,y2,"R",linewidth=10, label="Line Two")

p.title("Epic Info")
p.xlabel("X axis")
p.ylabel("Y axis")
p.legend()
p.grid(True,color="b")
p.show()
```
